```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(data['average_score'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Average Scores')
plt.xlabel('Average Score')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

The histogram illustrates a normal distribution of average scores, centered around the 50-unit mark. The majority of students fall within the 40-60 range, indicating a consistent performance across the dataset with minimal outliers at the extreme ends of the spectrum.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter students with average scores higher than 50
high_avg_score_students = data[data['average_score'] > 50]

# Calculate the count of students for each gender in this subset
gender_high_avg_count = high_avg_score_students.groupby('gender').size().reset_index(name='count_of_students')

# Create the bar chart
plt.figure(figsize=(8, 6))
sns.barplot(x='gender', y='count_of_students', data=gender_high_avg_count, hue='gender', legend=False, palette='viridis')
plt.title('Count of Students with Average Score > 50 by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Students')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

The bar chart demonstrates a nearly equal distribution between female and male students achieving an average score above 50. While male students show a slightly higher count, the minimal difference indicates a balanced academic performance across both genders within this performance bracket.

```python
high_avg_score_students = data[data['average_score'] > 40]
test_prep_high_score_counts = high_avg_score_students['test_preparation_course'].value_counts().reset_index()
test_prep_high_score_counts.columns = ['test_preparation_course', 'count_of_students']

plt.figure(figsize=(8, 6))
sns.barplot(x='test_preparation_course', y='count_of_students', data=test_prep_high_score_counts, hue='test_preparation_course', legend=False, palette='viridis')
plt.title('Students with Average Score > 60 by Test Preparation Course')
plt.xlabel('Test Preparation Course')
plt.ylabel('Number of Students')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

The bar chart shows that a significantly higher number of students who achieved good scores did not complete a formal test preparation course. While the "uncomplete" group is larger in volume (exceeding 5000 students), this highlights the importance of analyzing the success rate percentage relative to the total population of each group to fully understand the course's impact

```python
high_avg_score_students = data[data['average_score'] > 80]
parental_education_counts = high_avg_score_students['parental_education'].value_counts().reset_index()
parental_education_counts.columns = ['parental_education', 'count']

plt.figure(figsize=(10, 6))
sns.barplot(x='parental_education', y='count', data=parental_education_counts, hue='parental_education', legend=False, palette='viridis')
plt.title('Count of Students with Average Score > 60 by Parental Education')
plt.xlabel('Parental Education')
plt.ylabel('Number of Students')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

The bar chart indicates that students whose parents hold a Bachelor’s degree represent the highest count among top performers (scores > 80). Interestingly, the PhD category shows the lowest count, which likely reflects the smaller population of PhD holders in the dataset rather than a lack of academic influence. Overall, a strong academic background at home correlates with high student achievement
